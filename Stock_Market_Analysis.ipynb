### Plot Moving Averages for ITC

#### In an average calculation where the latest data is included, and the oldest is excluded called a Moving Average. The reason for calculating the moving average of a stock is to help smooth out the price data over a specified period of time by creating a constantly updated average price.

```python
itc['Open'].plot()
itc['MA50'] = itc['Open'].rolling(50).mean()
itc['MA50'].plot(label='MA50')
itc['MA200'] = itc['Open'].rolling(200).mean()
itc['MA200'].plot(label='MA200')
```

## Daily Percentage Change

### To find volatility of the stock and if the stock is stable

```python
tcs['returns'] = (tcs['Close'] / tcs['Close'].shift(1)) - 1
itc['returns'] = (itc['Close'] / itc['Close'].shift(1)) - 1
infy['returns'] = (infy['Close'] / infy['Close'].shift(1)) - 1
wipro['returns'] = (wipro['Close'] / wipro['Close'].shift(1)) - 1
```

```python
tcs.head()
```

```python
# Replace NA values with 0
tcs['returns'].fillna(0 , inplace=True)
infy['returns'].fillna(0 , inplace=True)
itc['returns'].fillna(0 , inplace=True)
wipro['returns'].fillna(0 , inplace=True)
```

### Plotting the returns on historgram

```python
import seaborn as sns
sns.set()
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(2, 2, figsize=(12, 7))
sns.distplot(tcs['returns'] , color="b", ax=axes[0, 0], axlabel='TCS');
sns.distplot(infy['returns'] , color="r", ax=axes[0, 1], axlabel='INFY');
sns.distplot(itc['returns'] , color="g", ax=axes[1, 0], axlabel='ITC');
sns.distplot(wipro['returns'] , color="m", ax=axes[1, 1], axlabel='WIPRO');
```
